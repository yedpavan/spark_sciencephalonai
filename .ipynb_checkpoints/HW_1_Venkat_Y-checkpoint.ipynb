<b>Transformations and Actions:</b><br>
Spark creates only logic and creates an RDD from existing RDD when Transformations are applied. It executes the logic only when Action is called.<br><br>
<b>Lazy Evaluation:</b><br>
spark does not execute transformations immediately instead it  
1. builds logic <br>
2. optimizes the execution <br>
3. Executes it only when the Actions are called.<br>

<b>Benefits</b>
1. unused transformations are not executed; saves computations and memory
2. can combine multiple transformations

### Question#5

<b>reduceByKey:</b><br>
It operates in two stages <br>
a. Combine values with the same key within each partition<br>
b. Combine results from all partitions for each key<br>
Internally It works as : 
1. Spark partitions the data across the cluster.
2. Each partition combines values with the same key locally.
3. It then shuffles the data, ensuring all values for a key are on the same node so that reduction function (sum) can be applied to entire dataset
4. Finally, it combines the results for each key.

<b>Mapreduce</b> is a programming model that uses parallel processing to process large amounts of data.It works in 3 steps:
1. Map: Input data is split into subsets and converted into key-value pairs. 
2. Shuffle: Key-value pairs are grouped by key and redistributed so that all data with the same key is on the same node. 
3. Reduce: Values associated with the same key are aggregated and processed to produce a final set of key-value pairs.
<br>

<b>flatMap->map->reduceByKey </b>  is a common pattern in distributed data processing similar to Mapreduce<br>
1. flatmap() is used to split line into words and then emit each word as a key-value pair
2. map() is used to convert each word into a key_value pair of the form (word,1)
3. reduceByKey is used to combine the values for each key after the data has been shuffled.

# Exercise_5: Spark Application Architecture

### Question#1: Spark execution model

<b>Driver</b> is the heart of a Spark application which maintains all relevant information during the lifetime of the application.It is responsible for breaking the job into stages and scheduling tasks for execution. <br>
<b>Executors</b> are distributed worker nodes that run these tasks in parallel and store data if caching is enabled. Each executor is responsible for only two things: executing code assigned to it by the driver and reporting the state of the computation, on that executor, back to the driver node.<br>
<b>Ref</b>: https://www.databricks.com/glossary/what-are-spark-applications#:~:text=The%20driver%20process%20is%20absolutely,that%20the%20driver%20assigns%20them.

### Question#2: Jobs and Actions

<b>transformations</b> are operations that define how to manipulate data by creating new datasets based on existing ones without immediately executing the changes<br>
<b>actions</b> trigger the execution of those transformations and return a result and <br>
<b>jobs</b> represents a complete set of operations (transformations and actions) that Spark needs to perform to achieve a specific task, essentially a sequence of transformations initiated by an action. <b>So Action triggers the jobs</b><br>

So, if 2 actions are called in our code, 2 jobs will be launched<br>
 df.count() # Action 1 triggers job 1<br>
 df.show() # Action 2 triggers job 2<br>


<b>Ref</b>: https://medium.com/@roshmitadey/pyspark-transformations-v-s-actions-797fc8ad16ea

### Question#4

<b>Efficiency:Lazy Evaluation</b><br>
spark analyses the entire transformation and pipeline them; if transformations are applied multiple times, spark avoids unnecessary processing.<br>
<b>Fault Tolerance : <b><br>
1. Spark constructs a DAG of transformations, breaking the job into stages
2. If an action fails, Spark can recompute the failed or lost partition from DAG transformation instead of restarting the entire job

<b>Importance of DAG</b>
1. Spark performs the jobs one after the other without DAG, which is inefficient way of processing
2. Fault recovery
3. Optimized execution by narrow transformation and reduce shuffling
